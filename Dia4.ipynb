# Tercer Challenge
Crear una agenda de contactos, el usuario puede:
1- Mostrar los contactos
2- Crear contacto
3- Modificar contacto

```python
guia_telefonica = {}

while True:
    accion = input('Guia telefonica\n Presione 1 para agregar un contacto\n Presione 2 para modificar contacto' \
    'Presione 3 para eliminar contacto\n Presione 4 para ver la lista de contactos\n Presione 5 e indique el contacto para ver el numero'
    '\nPresione otro para terminar')
    
    if accion == '1':
        contacto = input('Nombre y apellido')
        numero = int(input('Numero de telefono'))
        guia_telefonica[contacto]=numero
    elif accion == '2':
        print('Indique que contacto le gustaria modificiar')
        guia_telefonica.keys()
        contacto = input('Cambiar nombre')
        numero2 = input('Nuevo numero')
        guia_telefonica[contacto]= numero2
    elif accion == '3':
        print('Indique e contacto a eliminar')
        guia_telefonica.keys()
        contacto = input('')
        del guia_telefonica[contacto]
    elif accion == '4':
        print(guia_telefonica)
    elif accion == '5':
        contacto = input('')
        print(f'{guia_telefonica[contacto]}')
    else:
        break

```

# Quinto Challenge

Crear 2 metodos para nestra mascota y ejecutar, debe ser algo especifico de esa mascota.
'Ejemeplo: "Cuando hablo, digo '...'. Cuando me muevo, hago'...'"


```python
#Creamos la clase
class Felino:
    #Definimos atributo de clase
    especie = 'gato'
    #Creamos atributos de instancia
    def __init__(self,nombre,color):
        self.nombre = nombre
        self.color = color
    def sonido(self, sound_effect):
            self.sound = sound_effect
            print(f'El {self.especie} hace {sound_effect}')
    def movimiento(self, caminar):
            self.caminar = caminar
            print(f'El {self.especie} camina {caminar}')

felino1 = Felino('Cucho','Blanco')
felino1.sonido('miau')
felino1.movimiento('modelando')
```

# Sexto Challenge

```python
#Clase MADRE
class Mage:
    hp = 15
    magic = 12
    #Aributos de instancia
    def __init__(self, atk1, atk2 ):
        self.atk1 = atk1
        self.atk2 = atk2
    #Acciones de clase
    def ataque1(self):
        print(f'I cast {self.atk1}')
    def ataque2(self):
        print(f'I cast {self.atk2}')


#Clase Hija
class DarkMage(Mage):
    pass
#Definimos los atributos de instancia para un elemento de la clase hija
mago1 = DarkMage('Fire strike','Blood Barrage')
#Realizamos una accion de la clase madre con un elemento de la clase hija
mago1.ataque2()

```
