###CUARTO CHALLENGE
### Crea una funcion que acepte 2 parametros(palabra y numero) y que retorne TRUE si la cantidad de letras es igual al numero, y FALSE en caso contrario.

1- Hacemos 2 variables con input, una para la palabra, otra para el numero(incluir en un int()).
2- Definimos la funcion con 2 parametros
3- Contamos la cantidad de letras con "len()"
4- Creamos un "if" y comparamos el "len(palabra)" con el numero
5- Condicionamos los return
6- Llamamos a la funcion

```python
palabra_a_contar = input('Ingrese la palabra')
cantidad_de_letras = int(input('Cantidad de letras'))

def contar_letras(palabra, numero):
    if len(palabra)==numero:
        return True
    else:
        return False

contar_letras(palabra_a_contar, cantidad_de_letras)
```

### QUINTO CHALLENGE
### Crae una funcion que revise si una palabra contiene una letra. En caso de que la contenga, retornar la letra entre [], en caso de que no retornar la letra.

```python
def verificar_letra():
    palabra = input('Ingrese una palabra')
    letra = input('Verificar si la palabra tiene la letra')
    if letra in palabra:
        return [letra]
    else:
        return letra
    
verificar_letra()
```
