```python
temperatura = int(input('Ingrese la temperatura en ºC'))

print(f'{temperatura}ºC')

def como_esta_el_clima():
    if temperatura >= -10 and temperatura <= 18:
        print('Fresquete')
    elif temperatura >= 19 and temperatura <= 30:
        print("Calor'i")
    elif temperatura >= 31 and temperatura <= 45:
        print("Hakuuuuu")
    else:
        print("Ñamano mba'e!")

como_esta_el_clima()
```

```python
def como_esta_el_dia():
    temp = int(input('Que temperatura hace?'))
    print(f'Hace {temp}ºC')
    if temp >= -10 and temp <= 18:
        print('Fresquete')
    elif temp >= 19 and temp <= 30:
        print("Calor'i")
    elif 31 <= temp and temp <= 45:
        print("Hakuuuuu")
    else:
        print("Ñamano mba'e!")
como_esta_el_dia()
```
